```python
df_end=df.groupby("Destination Airport").agg(Landings=("Origin Airport","count")).sort_values(
    by="Landings",ascending=False).reset_index()
df_end=df_end.rename(columns={"Destination Airport":"Airport"})
df_end.head(15)
```

```python
df_start=df.groupby("Origin Airport").agg(Starts=("Destination Airport","count")).sort_values(
    by="Starts",ascending=False).reset_index()
df_start=df_start.rename(columns={"Origin Airport":"Airport"})
df_start.head(15)
```

```python
df_air=df_end.merge(df_start,how="outer",on="Airport")
df_air["Starts"]=df_air["Starts"].fillna(0)
df_air["Landings"]=df_air["Landings"].fillna(0)
df_air["Flight Movements"]=df_air["Starts"]+df_air["Landings"]
df_air.sort_values(by="Flight Movements",ascending=False).reset_index().head(20)
```
