```python
def place_is_home(date_end,place):
    home_flag=0
    if date_end<datetime.datetime(2006, 10, 17): 
        if place == "Unterlauchringen":
            home_flag=1
    elif date_end<datetime.datetime(2007, 3, 16):
        if place == "Meyrin":
            home_flag=1
    elif date_end<datetime.datetime(2007, 5, 17):
        if place == "Unterlauchringen":
            home_flag=1
    elif date_end<datetime.datetime(2020, 8, 29):
        if place == "Genf":
            home_flag=1
    elif date_end<datetime.datetime(2021, 4, 29):
        if place == "Unterlauchringen":
            home_flag=1
    elif date_end<(datetime.date.today()+datetime.timedelta(days=7)):
        if place == "Wien":
            home_flag=1
    return home_flag
```
