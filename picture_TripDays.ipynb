```python
def func_cot(x_list):
    xl_r=[]
    for x_l in x_list:
        if len(x_l)>len(xl_r):
            xl_r = x_l
    return xl_r
```

```python
df_place_agg=df.loc[(df["Place"]==df['Trip Origin'])].groupby(["Date","Place"]).agg( 
                        DayTrip=('Day-Trip','min'),
                        PictureCountAtPlace=("Picture Count",'sum'),
                        CoTravellers=("Co-Travellers",lambda x: func_cot(x)),
                        NoCoT=("No Co-Travellers",'max')
).reset_index()
df_place_agg
```
