```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("~/jupytermacros/datasets/PhotosPerCountryPerYear.csv")
df.set_index('year',inplace=True)
df_cum=df.cumsum(axis=0)
df_cum.to_csv("~/jupytermacros/datasets/PhotosPerCountryCumultativeAfterYear.csv")
df_grouped=df.copy()
#print("before orig-replaced", df.sum(axis=1)-df_grouped.sum(axis=1))
df_grouped["other European Countries"]=(df_grouped["Norway"]+df_grouped["Luxembourg"]+
        df_grouped["Russia"]+df_grouped["Denmark"]+df_grouped["Belgium"]+df_grouped["Portugal"]+
        df_grouped["Turkey"]+df_grouped["Vatican"]+df_grouped["Sweden"]+df_grouped["Netherlands"]+
        df_grouped["Poland"]+df_grouped["Hungary"]+df_grouped["Czech"]++df_grouped["Iceland"]+
        df_grouped["Greece"]+df_grouped["Ukraine"]+df_grouped["Croatia"]+df_grouped["Monaco"]+
        df_grouped["Slovakia"]+df_grouped["Estonia"]+df_grouped["Ireland"]+df_grouped["Finland"]+
        df_grouped["Slovenia"]+df_grouped["San Marino"]+df_grouped["Liechtenstein"]+df_grouped["Serbia"]+
        df_grouped["Romania"]+df_grouped["Latvia"]+df_grouped["Lithuania"]+df_grouped["North Macedonia"]+
        df_grouped["Malta"])                                
df_grouped["Africa"]=df_grouped["Tunisia"]+df_grouped["Egypt"]+df_grouped["Morocco"]
df_grouped["Asia"]=df_grouped["China"]+df_grouped["South Korea"]+df_grouped["Japan"]+df_grouped["North Korea"]+df_grouped["Pakistan"]+df_grouped["Iraq"]+df_grouped["UAE"]+df_grouped["Saudi-Arabia"]+df_grouped["Qatar"]+df_grouped["Kuwait"]
df_grouped["USA"]=df_grouped["USA"]+df_grouped["Canada"]
#print(df_grouped.head())
#print("after1 orig-replaced", df.sum(axis=1)-df_grouped.sum(axis=1))
#drop original columns africa and asia first
df_grouped=df_grouped.drop(
    ["Tunisia","Egypt","Morocco","China","South Korea","Japan","North Korea","UAE","Iraq",
     "Pakistan","Saudi-Arabia","Qatar","Kuwait"],axis=1)
#then drop european other countries columns (per default tries dropping rows, aka axis=0)
df_grouped=df_grouped.drop(
    ["Russia","Denmark","Belgium","Portugal","Turkey","Vatican","Sweden","Netherlands","Poland",
     "Hungary","Czech","Iceland","Greece","Ukraine","Croatia","Monaco","Slovakia","Estonia",
     "Ireland","Finland","Slovenia","San Marino","Liechtenstein","Serbia","Romania","Bulgaria",
     "Norway","Luxembourg",'Lithuania','Latvia','North Macedonia','Malta'],axis=1)
#so few candian photos, add them to US
df_grouped=df_grouped.drop(["Canada"],axis=1)
df_grouped_cum=df_grouped.cumsum(axis=0)
print("tail groups/summed group",df_grouped.tail(),df_grouped_cum.tail())
#print("after2 orig-replaced", df.sum(axis=1)-df_grouped.sum(axis=1))
ax = df_grouped.plot(kind='bar', stacked=True,width=1,figsize=(18.5, 10.5))
ax.set_ylabel('photos per year', fontsize=35)
ax.set_xlabel('year', fontsize=25)
#legend_params = {'legend.fontsize': 80,
#          'legend.handlelength': 2}
#plt.rcParams.update(legend_params)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.rc('legend',fontsize=30, frameon=False) # using a size in points
plt.savefig('Photos_PerYear_and_Country_absolute.png')
plt.show()

ax_cum = df_grouped_cum.plot(kind='bar', stacked=True,width=1,figsize=(18.5, 10.5))
ax_cum.set_ylabel('total photos after year', fontsize=35)
ax_cum.set_xlabel('year', fontsize=25)
#legend_params = {'legend.fontsize': 80,
#          'legend.handlelength': 2}
#plt.rcParams.update(legend_params)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.rc('legend',fontsize=25, frameon=False) # using a size in points
plt.savefig('Photos_PerYear_and_Country_absoluteCumultative.png')
plt.show()

df_new=df_grouped.div(df_grouped.sum(axis=1), axis=0)
#df_new.head()
ax2 = df_new.plot(kind='bar', stacked=True,width=1,figsize=(18.5, 10.5))
ax2.set_ylabel('fraction of photos per year [in %]', fontsize=25)
ax2.set_xlabel('year', fontsize=25)
#box definition: first digi is x value of left size
plt.rc('legend',fontsize=20, frameon=False) # using a size in points
#ax2.legend(loc='upper center', bbox_to_anchor=(0.5, 1.50),ncol=3, fancybox=True, shadow=True)
ax2.legend(loc='lower left', bbox_to_anchor=(0.0,1.02, 1., .102),ncol=3, mode="expand", borderaxespad=0.,fancybox=True, shadow=True)
#legend_params = {'legend.fontsize': 80,
#          'legend.handlelength': 2}
#plt.rcParams.update(legend_params)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.tight_layout()
plt.savefig('Photos_PerYear_and_Country_fraction.png')
plt.show()

df_new_cum=df_grouped_cum.div(df_grouped_cum.sum(axis=1), axis=0)
#df_new.head()
ax2_cum = df_new_cum.plot(kind='bar', stacked=True,width=1,figsize=(18.5, 10.5))
ax2_cum.set_ylabel('fraction of photos after year [in %]', fontsize=25)
ax2_cum.set_xlabel('year', fontsize=25)
#box definition: first digi is x value of left size
plt.rc('legend',fontsize=20, frameon=False) # using a size in points
#ax2.legend(loc='upper center', bbox_to_anchor=(0.5, 1.50),ncol=3, fancybox=True, shadow=True)
ax2_cum.legend(loc='lower left', bbox_to_anchor=(0.0,1.02, 1., .102),ncol=3, mode="expand", borderaxespad=0.,fancybox=True, shadow=True)
#legend_params = {'legend.fontsize': 80,
#          'legend.handlelength': 2}
#plt.rcParams.update(legend_params)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.tight_layout()
plt.savefig('PhotosCumulatatedAfterYear__fraction.png')
plt.show()
```
